```python
import pandas as pd
```

```python
df = pd.read_excel("API_BEN_DS2.xls", skiprows=range(3))
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.drop(['Country Name', 'Country Code', 'Indicator Code'], axis=1, inplace=True)
```

```python
df.rename(columns={'Indicator Name' : 'Indicateur'}, inplace=True)
```

```python
df.set_index('Indicateur', inplace=True)
```

```python
df.head(10)
```

```python
years = list(map(str, range(1960, 2019)))
years
```

```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
imp = df.loc['Merchandise imports (current US$)', years]
imp
```

```python
imp.plot(kind='area')
plt.title('Merchandise imports (current US$)')
plt.ylabel('Amount (USD)')
plt.xlabel('Years')

plt.show()
```

```python
food = df.loc['Food imports (% of merchandise imports)', years]
```

```python
mean = food.mean()
```

```python
import numpy as np
food.replace(np.nan, mean, inplace=True)
```

```python
food.plot(kind='line')
plt.title('Food imports (% of merchandise imports)')
plt.ylabel('Percentage of merchandise import')
plt.xlabel('Years')
plt.show()
```
